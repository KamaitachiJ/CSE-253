```python
def addBias(X):
    X = np.insert(X, 0, 1, axis=0)
    return X
```

```python
trainLabel = np.array(trainLabel).T
testLabel = np.array(testLabel).T
trainData = np.asarray(trainData).T / 127.5 - 1
testData = np.asarray(testData).T / 127.5 - 1
testData = addBias(testData)
```

## Basic Test

```python
from sklearn.model_selection import train_test_split

def sigmoid(x):
    return 1.0 / (1 + np.exp(-x))

def tanh(x):
    return 1.7159 * np.tanh(2. * x / 3)

def ReLU(x):
    return x * (x > 0)

def addBias(data):
    bias = np.ones(len(data))[:, np.newaxis]
    biasData = np.concatenate((bias,data), axis=1)
    return biasData

def oneHot(label):
    labelHot = np.zeros((len(label),len(np.unique(label))))
    for i in range(len(label)):
        labelHot[i,label[i]] = 1
    return labelHot

def calculateLoss(data, label, w1, w2, lam, actFun):
    a_j = data.dot(w1)
    if actFun == "sigmoid":
        ga_j = sigmoid(a_j)
    elif actFun == "tanh":
        ga_j = tanh(a_j)
    elif actFun == "ReLU":
        ga_j = ReLU(a_j)
    a_k = addBias(ga_j).dot(w2)
    ga_k = np.exp(a_k) / np.sum(np.exp(a_k), axis=1)[:, np.newaxis]
    E = -1.0 * np.mean(label * np.log(ga_k))
    C = 0.5 * (np.sum(np.power(w1, 2)) + np.sum(np.power(w2, 2)))
    loss = E + lam * C
    #loss = E
    return loss

def errRate(data, label, w1, w2, actFun):
    a_j = data.dot(w1)
    if actFun == "sigmoid":
        ga_j = sigmoid(a_j)
    if actFun == "tanh":
        ga_j = tanh(a_j)
    elif actFun == "ReLU":
        ga_j = ReLU(a_j)
    a_k = addBias(ga_j).dot(w2)
    ga_k = np.exp(a_k) / np.sum(np.exp(a_k), axis=1)[:, np.newaxis]
    prediction = np.argmax(ga_k, axis = 1)
    err = 1.0 * sum(a != b for (a,b) in zip(prediction, label)) / len(label)
    return err

def oneHiddenLayer(trainData, trainLabel, testData, testLabel, lr = 1e0, maxIter = 100, T = 2e8, lam = 1e-5, \
                   batchSize = 128, unitNum = 64, shuffle = False, actFun = "sigmoid", randInitial = False, mu = 0, fitOnly = False):
    # tack on a "1" at the beginning for a bias parameter
    trainData = addBias(trainData)
    if not fitOnly:
        trainLoss = []
        validLoss = []
        testLoss = []
        trainErr = []
        validErr = []
        testErr = []
    
    classNum = len(np.unique(trainLabel))
    
    trainD, validD, trainL, validL = train_test_split(trainData, trainLabel, test_size=0.1, random_state=42)    
    
    trainLHot = oneHot(trainL)
    if randInitial:
        w_ij = np.random.normal(0,1. / np.sqrt(len(trainD[0])),(len(trainD[0]), unitNum))
        w_jk = np.random.normal(0,1. / np.sqrt(unitNum + 1),(unitNum + 1, classNum))
    else:
        w_ij = np.random.normal(0,1,(len(trainD[0]), unitNum))
        w_jk = np.random.normal(0,1,(unitNum + 1, classNum))
    if shuffle:
        index = np.arange(len(trainD))
    if mu != 0:
        v_ij = 0.
        v_jk = 0.
    it = 0
    
    while it <= maxIter:
        if shuffle:
            np.random.shuffle(index)
        lr1 = lr / (1.0 + it / T)
        if mu != 0:
            mu1 = mu * (1.0 + it / (2*T))
            if mu1 > 0.9:
                mu1 = 0.9
        # Mini Batch Gradient Descent
        for startInd in range(0, len(trainData) - batchSize + 1, batchSize):
            if shuffle:
                trainDBatch = trainD[index[startInd : startInd + batchSize]]
                trainLBatch = trainLHot[index[startInd : startInd + batchSize]]
            else:
                trainDBatch = trainD[startInd : startInd + batchSize]
                trainLBatch = trainLHot[startInd : startInd + batchSize]
            
            #Forward Propagation
            #1. Input to Hidden Layer
            a_j = trainDBatch.dot(w_ij)   #(batchSize, unitNum) 
            if actFun == "sigmoid":
                ga_j = addBias(sigmoid(a_j))   #(batchSize, unitNum + 1)
            elif actFun == "tanh":
                ga_j = addBias(tanh(a_j))
            elif actFun == "ReLU":
                ga_j = addBias(ReLU(a_j))
            
            #2. Hidden Layer to Output
            a_k = ga_j.dot(w_jk)   #(batchSize, classNum)
            ga_k = np.exp(a_k) / np.sum(np.exp(a_k), axis=1)[:, np.newaxis]   #(batchSize, classNum)
            
            #Back Propagation
            delta_k = ga_k - trainLBatch   #(batchSize, classNum)
            if actFun == "sigmoid":
                derGa_j = ga_j * (1 - ga_j)   #(batchSize, unitNum)
            elif actFun ==  "tanh":
                derGa_j = (2. / 3) * 1.7159 * (1 - np.power(ga_j / 1.7159, 2))
            elif actFun == "ReLU":
                derGa_j = (1. * (ga_j > 0))
            delta_j = (derGa_j * (delta_k.dot(w_jk.T)))[:,1:]   #(batchSize, unitNum)
            
            # Update weights
            dw_jk = lr1 * (ga_j.T.dot(delta_k) / batchSize)
            dw_ij = lr1 * (trainDBatch.T.dot(delta_j) / batchSize)
            if mu != 0:
                v_jk = mu1 * v_jk - lr1 * dw_jk
                v_ij = mu1 * v_ij - lr1 * dw_ij
                w_jk = w_jk + v_jk - np.concatenate((np.zeros((1,len(w_jk[0]))),\
                                                    lam * w_jk[1:]), axis = 0)  #(unitNum + 1, classNum)
                w_ij = w_ij + v_ij - np.concatenate((np.zeros((1,len(w_ij[0]))), \
                                                      lam * w_ij[1:]), axis = 0)   #(featuresNum, unitNum)
            else:
                w_jk = w_jk - dw_jk - np.concatenate((np.zeros((1,len(w_jk[0]))),\
                                                      lam * w_jk[1:]), axis = 0)  #(unitNum + 1, classNum)
                w_ij = w_ij - dw_ij - np.concatenate((np.zeros((1,len(w_ij[0]))), \
                                                      lam * w_ij[1:]), axis = 0)   #(featuresNum, unitNum)
            if startInd % 12800 == 0 and not fitOnly:
                trainLoss.append(calculateLoss(trainD, trainLHot, w_ij, w_jk, lam, actFun))
                validLoss.append(calculateLoss(validD, oneHot(validL), w_ij, w_jk, lam, actFun))
                testLoss.append(calculateLoss(testData, oneHot(testLabel), w_ij, w_jk, lam, actFun))
                trainErr.append(errRate(trainD, trainL, w_ij, w_jk, actFun))
                validErr.append(errRate(validD, validL, w_ij, w_jk, actFun))
                testErr.append(errRate(testData, testLabel, w_ij, w_jk, actFun))
        lossV = calculateLoss(validD, oneHot(validL), w_ij, w_jk, lam, actFun)
        errV = errRate(validD, validL, w_ij, w_jk, actFun)
        if it % 10 == 0:
            print (lossV, errV)
        it += 1
    print ('done')
    if not fitOnly:
        return w_ij, w_jk, trainLoss, validLoss, testLoss, trainErr, validErr, testErr
    return w_ij, w_jk   
```

```python
w_ij, w_jk, trainLoss, validLoss, testLoss, trainErr, validErr, testErr = \
oneHiddenLayer(trainData.T, trainLabel, testData.T, testLabel)
```

```python
print np.max(1 - np.asarray(testErr))
x = [i for i in range(len(trainLoss))]
plt.figure(figsize = (8,6))
trainL, = plt.plot(x, trainLoss)
validL, = plt.plot(x, validLoss)
testL, = plt.plot(x, testLoss)
plt.legend([trainL, validL, testL], ['trainLoss', 'validLoss', 'testLoss'])
plt.grid(True)
plt.xlabel('# of iteration')
plt.ylabel('Loss')
plt.title('Loss of Basic Test')
x = [i for i in range(len(trainErr))]
plt.figure(figsize = (8,6))
trainE, = plt.plot(x, 1-np.asarray(trainErr))
validE, = plt.plot(x, 1-np.asarray(validErr))
testE, = plt.plot(x, 1-np.asarray(testErr))
plt.legend([trainE, validE, testE], ['trainRate', 'validRate', 'testRate'])
plt.grid(True)
plt.xlabel('# of iteration')
plt.ylabel('Percent of accuracy')
plt.title('Percent of accuracy of Basic Test')
plt.show()
```
